```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Se carga el archivo csv

```python
amazon_prime_tittles = pd.read_csv("amazon_prime_titles.csv")
```

Se observa la informacion que contiene el csv

```python
amazon_prime_tittles.info
```

Se crea una columna llamada "id", la cual contiene la letra inicial de la plataforma junto con el "show id". Por ejemplo la plataforma Amazon y el show id s1, la columna id sería igual a as1.

```python
amazon_prime_tittles["id"] = "a" + amazon_prime_tittles["show_id"]
amazon_prime_tittles
```

Todos los valores nulos de la columna  rating se renombran con la letra "G"

```python
amazon_prime_tittles["rating"].fillna("G", inplace = True)
amazon_prime_tittles
```

Se cambia el formato de la columna date added al formato año/mes/dia

```python
amazon_prime_tittles["date_added"] = pd.to_datetime(amazon_prime_tittles["date_added"])
amazon_prime_tittles
```

Se convierten en minusculas todas las palabras contenidas en las columnas: type, title, director, cast, country, date addes, listed in, description.

```python
amazon_prime_tittles["type"] = amazon_prime_tittles["type"].str.lower()
amazon_prime_tittles["title"] = amazon_prime_tittles["title"].str.lower()
amazon_prime_tittles["director"] = amazon_prime_tittles["director"].str.lower()
amazon_prime_tittles["cast"] = amazon_prime_tittles["cast"].str.lower()
amazon_prime_tittles["country"] = amazon_prime_tittles["country"].str.lower()
amazon_prime_tittles["listed_in"] = amazon_prime_tittles["listed_in"].str.lower()
amazon_prime_tittles["description"] = amazon_prime_tittles["description"].str.lower()
```

Se separan en 2 la columna duration. Por un lado una columna contiene el tipo de dato (min, sesion) y por otro lado el numero asociado a la duracion

```python
name=amazon_prime_tittles["duration"].str.split(expand=True)
name.columns=["duration_int", "duration_type"]
amazon_prime_tittles=pd.concat([amazon_prime_tittles, name], axis=1)
```

Se eliminda la columna duracion, pues ya quedo separada

```python
amazon_prime_tittles.drop(["duration"], axis=1, inplace=True)
```

Se crea la columna respectiva para cada plataforma

```python
amazon_prime_tittles["Plataforma"] = "amazon"
amazon_prime_tittles
```

Se realizan todas las transformaciones para todos los csv

ETL del csv Netflix

```python
netflix_titles =pd.read_csv("netflix_titles (1).csv")
netflix_titles["id"] = "n" + netflix_titles["show_id"]
netflix_titles["rating"].fillna("G", inplace = True)

netflix_titles["type"] = netflix_titles["type"].str.lower()
netflix_titles["title"] = netflix_titles["title"].str.lower()
netflix_titles["director"] = netflix_titles["director"].str.lower()
netflix_titles["cast"] = netflix_titles["cast"].str.lower()
netflix_titles["country"] = netflix_titles["country"].str.lower()
netflix_titles["date_added"] = netflix_titles["date_added"].str.lower()
netflix_titles["listed_in"] = netflix_titles["listed_in"].str.lower()
netflix_titles["description"] = netflix_titles["description"].str.lower()

name=netflix_titles["duration"].str.split(expand=True)
name.columns=["duration_int", "duration_type"]
netflix_titles=pd.concat([netflix_titles, name], axis=1)
netflix_titles.drop(["duration"], axis=1, inplace=True)
netflix_titles["date_added"] = pd.to_datetime(netflix_titles["date_added"])
netflix_titles["Plataforma"] = "netflix"
```

ETL del csv Disney

```python
disney_plus_titles = pd.read_csv("disney_plus_titles.csv")

disney_plus_titles["id"] = "d" + disney_plus_titles["show_id"]
disney_plus_titles["rating"].fillna("G", inplace = True)

disney_plus_titles["type"] = disney_plus_titles["type"].str.lower()
disney_plus_titles["title"] = disney_plus_titles["title"].str.lower()
disney_plus_titles["director"] = disney_plus_titles["director"].str.lower()
disney_plus_titles["cast"] = disney_plus_titles["cast"].str.lower()
disney_plus_titles["country"] = disney_plus_titles["country"].str.lower()
disney_plus_titles["date_added"] = disney_plus_titles["date_added"].str.lower()
disney_plus_titles["listed_in"] = disney_plus_titles["listed_in"].str.lower()
disney_plus_titles["description"] = disney_plus_titles["description"].str.lower()

name=disney_plus_titles["duration"].str.split(expand=True)
name.columns=["duration_int", "duration_type"]
disney_plus_titles=pd.concat([disney_plus_titles, name], axis=1)

disney_plus_titles.drop(["duration"], axis=1, inplace=True)

disney_plus_titles["date_added"] = pd.to_datetime(disney_plus_titles["date_added"])
disney_plus_titles["Plataforma"] = "disney"

```

ETL del csv Hulu

```python
hulu_titles = pd.read_csv("hulu_titles.csv")

hulu_titles["id"] = "h" + hulu_titles["show_id"]
hulu_titles["rating"].fillna("G", inplace = True)

hulu_titles["type"] = hulu_titles["type"].str.lower()
hulu_titles["title"] = hulu_titles["title"].str.lower()
hulu_titles["director"] = hulu_titles["director"].str.lower()
#hulu_titles["cast"] = hulu_titles["cast"].str.lower()
hulu_titles["country"] = hulu_titles["country"].str.lower()
hulu_titles["date_added"] = hulu_titles["date_added"].str.lower()
hulu_titles["listed_in"] = hulu_titles["listed_in"].str.lower()
hulu_titles["description"] = hulu_titles["description"].str.lower()

name= hulu_titles["duration"].str.split(expand=True)
name.columns=["duration_int", "duration_type"]
hulu_titles=pd.concat([hulu_titles, name], axis=1)
hulu_titles.drop(["duration"], axis=1, inplace=True)

hulu_titles["date_added"] = pd.to_datetime(hulu_titles["date_added"])
hulu_titles["Plataforma"] = "hulu"
```

Se juntan todos los archivos csv despues de realizarles a cada uno un ETL, en un solo archivo que contenga toda la información, para posterior uso de la data

```python
bases_datos = pd.concat([amazon_prime_tittles, disney_plus_titles, netflix_titles, hulu_titles])
bases_datos
```

Se exporta el dataframe en csv, una vez realizada todas las tranformaciones solicitadas.
