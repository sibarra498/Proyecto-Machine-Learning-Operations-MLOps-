Definimos como metrica, que si el Score de la pelicula es menor o igual a 3.2 no la recomendara, por el contrario si el Score es igual o mayor a 3.3 si la recomendara al usuario.

```python
conditions = [(bd["Score"] <= 3.2), 
               (bd["Score"]>= 3.3)]
values = ["No recomendado","Recomendado"]
bd["clasificacion pelicula"] = np.select(conditions, values)
```

Nos quedamos con los campos que vamos a utilizar

```python
bd = bd[["userId", "Score", "type", "title", "listed_in", "id", "Plataforma", "clasificacion pelicula"]]
```

Se cambian todos los tipos de datos de categoricos a numericos

```python
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder()

bd_train = pd.DataFrame(index = bd.index)


for label in bd.columns:
    print(label + "-encoded")
    #label = "Severity of Illness-encoded"
    encoder.fit(bd[[label]])
    bd_train[label + "-encoded"] = encoder.transform(bd[[label]])

display(bd)
display(bd_train)
```

Correlación de los datos

```python
def plot_corre_heatmap(corr):
    '''
    Definimos una función para ayudarnos a graficar un heatmap de correlación
    '''
    plt.figure(figsize=(12,10))
    sns.heatmap(corr, cbar = True,  square = False, annot=True, fmt= '.2f'
                ,annot_kws={'size': 15},cmap= 'coolwarm')
    plt.xticks(rotation = 45)
    plt.yticks(rotation = 45)
    # Arreglamos un pequeño problema de visualización
    b, t = plt.ylim() # discover the values for bottom and top
    b += 0.5 # Add 0.5 to the bottom
    t -= 0.5 # Subtract 0.5 from the top
    plt.ylim(b, t) # update the ylim(bottom, top) values
    plt.show()
```

```python
corr = bd_train.corr()
plot_corre_heatmap(corr)
```

Modelo de regresión logistica para predicción de datos

```python
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

Definimos la varible y como la etiqueta que queremos predecir, en donde se hace la separacion de los datos

```python
X = bd_train.drop(labels = ["clasificacion pelicula-encoded"], axis = 1) 
y = bd_train['clasificacion pelicula-encoded'] # Etiqueta a predecir
```

Se definen las variblaes ha utilizar, usando X y Y definidos anteriormente con el dataset

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
```

```python
y_pred
```

```python
print("Exactitud ", metrics.accuracy_score(y_test, y_pred))
```
